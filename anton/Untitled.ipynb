```python
T = 50
```

```python
def get_var(x):
    return ((x - x.sum()) ** 2).sum()
```

```python
var = np.array([get_var(sir[i:i + T]) for i in range(sir.shape[0] - T - 1)]) / 1000
```

```python
plt.plot(var)
plt.show()
```

```python
width = 2300
heigth = round(2300 * 3 / 4)
my_dpi = 300
plt.figure(figsize=(width/my_dpi, heigth/my_dpi), dpi=my_dpi)
plt.plot(var)
plt.savefig('var', dpi=my_dpi)
```

```python
def get_gist(a, b, n):
    return np.array([(j * (b - a) / n + a, (j + 1) * (b - a) / n + a) for j in range(n)], dtype=float)

def distr_density(var, N):
    f_T = np.zeros((N, ))
    gist = np.linspace(var.min(), var.max(), N + 1)
    for j in range(N):
        f_T[j] = np.sum(np.logical_and(var >= gist[j], var < gist[j + 1]))
    return f_T, gist
```

```python
f_T, gist = distr_density(var, 7)
f_T = f_T / T
```

```python
def get_bin(x):
    for i in range(gist.shape[0] - 1):
        if (gist[i] <= x <= gist[i + 1]):
            return i
```

```python
def cdf(x):
    j = get_bin(x)
    assert j is not None
    n = f_T.shape[0]
    if (j == n - 1):
        return 1
    else:
        return (n * x - j) * f_T[j + 1] + f_T[:j + 1].sum()
```

```python
F = np.zeros(var.shape[0])
var_sorted = np.linspace(var.min(), var.max(), var.shape[0])
for i, x in enumerate(var_sorted):
    F[i] = cdf(x)
    
```

```python
plt.plot(F)
plt.show()
```
