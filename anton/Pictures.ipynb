```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
x1 =np.random.hypergeometric(ngood=1002, nbad=290, nsample=1010, size=2000)  * 10 - 7500
plt.plot(x1)
plt.show()
```

```python
x2 = np.random.geometric(p=0.35, size=2000) * 10
plt.plot(x2)
plt.show()
```

```python
x3 = np.random.gamma(shape=1, size=1000) * 100
plt.plot(x3)
plt.show()
```

```python
x4 = np.random.laplace(10, 1, 1000) * 200
plt.plot(x4)
plt.show()
```

```python
x = np.concatenate([x4 * 0.5, x1, x3, x2 * 10, x4 * 0.6 - 100, x3 * 2 + 700])
plt.plot(x)
plt.show()
```

```python
def x_gen(a, b, c, d, e):
    x1 = np.random.hypergeometric(ngood=1002, nbad=290, nsample=1010, size=a)  * 10 - 7500
    x2 = np.random.geometric(p=0.35, size=b) * 10
    x3 = np.random.gamma(shape=1, size=c) * 100
    x4 = np.random.laplace(10, 1, d) * 200
    x = np.concatenate([x4 * 0.5, x1, x3, x2 * 10, x4 * 0.6 - 100, x3 * 2 + 700])
    return np.log(np.linspace(1, e, x.shape[0])) * x * np.random.normal(4000, 1, x.shape[0]) / 100000
```

```python
y1 = x_gen(1500, 1500, 1000, 1500, 10000) / 700
plt.plot(y1)
plt.show()
```

```python
y2 = x_gen(500, 2500, 500, 2000, 100) / 380
plt.plot(y2)
plt.show()
```
